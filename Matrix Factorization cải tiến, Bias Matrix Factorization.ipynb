# Bias Matrix Factorization

```python
import numpy as np
# np.random.seed(4)

# Bước 1. Tính nguy
def nguy():
    M =[]
    for i in range(0, A.shape[0]):
        for j in range(0, A.shape[1]):
            if A[i,j] !=0:
                M.append(A[i,j])
    print(M)
    nguy = np.mean(M)
    print('nguy:', nguy)
    return nguy
nguy = nguy()

# Bước 2. Tính bu
def bu():
    tu =0 # tong cac( rating- nguy) trên tử
    num =0 
    arr = []
    for i in range(0, A.shape[0]):
        for j in range(0, A.shape[1]):
            if(A[i,j]!=0):
                tu += (A[i,j] - nguy)
                num +=1
        arr.append( tu/num)
    return arr
bu = bu()
print('bu', bu)

#Bước 3. Tính bi
def bi():
    tu =0 # tong cac( rating- nguy) trên tử
    num =0 
    arr = []
    B = A.T
    for i in range(0, B.shape[0]):
        for j in range(0, B.shape[1]):
            if(B[i,j]!=0):
                tu += (B[i,j] - nguy)
                num +=1
        arr.append( tu/num)
    return arr
bi = bi()
print('bi', bi)
def MF_bias(A, K, nguy, bu, bi, eta=0.1,learning_rate =0.001, steps =500):
    # khởi tạo 2 ma trận W, H ngẫu nhiên
    W = np.random.rand(user, K)
    H = np.random.rand(K, item)
    step =0
    while(step < steps):
        for u in range(0, user):
            for i in range(0, item):
                if A[u][i] >0 :
                    r_bar =0
                    for k in range(K):
                        r_bar += np.dot(W[u][k], H[k][i])
                    pui_bar = nguy + bu[u]+ bi[i] + r_bar
                    eui = A[u][i] - pui_bar
                    
                    #Update weigth: nguy, bu, bi
                    nguy  += eta* eui
                    bu[u] += eta* (eui - learning_rate * bu[u])
                    bi[i] += eta* (eui - learning_rate * bi[i])
                    for k in range(K):
                        W[u][k] += 2*eta*(2*eui* H[k][i] - learning_rate*W[u][k])
                        H[k][i] += 2*eta*(2*eui* W[u][k] - learning_rate*H[k][i])
        step +=1
    return np.dot(W,H)

A = np.array([[1,4,5,0,3],
              [5,1,0,5,2],
              [4,1,2,5,0],
              [0,3,4,0,4]])
print('A', A)

user = A.shape[0] # số user
item = A.shape[1] # số item
print('So user:', user)
print('So item:', item)


MF_bias =MF_bias(A, 3, nguy, bu, bi, 0.01, 0.01, 1000)
MF_predict = MF_bias + nguy
Y =np.array([[round(MF_predict[i][j]) for j in range(item)]for i in range(user)]) #làm tròn
print('Ma tran ban dau:\n ', A)
print('Ma tran sau khi ap dung MF đã chuẩn hóa \n ', MF_bias)
print('Ma trận dự đoán, sau khi cộng thêm nguy:\n', MF_predict)
print('Ma tran làm tròn :\n', Y)
```
